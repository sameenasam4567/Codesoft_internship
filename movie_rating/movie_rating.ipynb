```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'D:\codesoft internship\movie_rating\IMDb Movies India.csv', encoding='ISO-8859-1')
```

```python
# Convert 'Duration' column to numeric, coercing errors to NaN
df['Duration'] = pd.to_numeric(df['Duration'], errors='coerce')

# Calculate the average duration, ignoring NaN values
average_duration = df['Duration'].mean()

# Fill missing values in 'Duration' column with the average duration
df['Duration'].fillna(average_duration, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
# Check if there are any missing values in the 'Rating' column
if df['Rating'].isnull().any():
    # Drop rows with missing values in the 'Rating' column
    df.dropna(subset=['Rating'], inplace=True)

```

```python
# Calculate the mean rating per year
yearly_mean_rating = df.groupby('Year')['Rating'].mean()

```

```python

# Check if there are any missing values in the calculated mean ratings
if yearly_mean_rating.isnull().any():
    # Drop rows with missing mean ratings
    yearly_mean_rating.dropna(inplace=True)

```
